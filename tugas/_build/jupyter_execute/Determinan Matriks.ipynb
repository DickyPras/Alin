# Determinan Matriks

## Definisi

Determinan matriks adalah selisih antara perkalian elemen-elemen pada diagonal utama dengan perkalian elemen-elemen pada diagonal sekunder. Determinan suatu matriks dapat dinotasikan dengan det(A) atau |A|.

Setiap matriks persegi ataupun bujur sangkar mempunyai nilai determinan. Nilai
determinan dari suatu matriks itu adalah skalar. Jika nilai determinan suatu matriks
sama dengan nol maka matriks tersebut disebut matriks singular.

```python
import numpy as np
a = np.array([[3,3], [5,2]])
b = np.array([[2,3,3],[4,5,1],[6,3,3]])
c = np.array([[1,0], [0,1]])
d = np.array([[1,0,0],[0,1,0],[0,0,1]])
print(a)
print()
print(b)
print()
print(c)
print()
print(d)
print()
print(f"determinan dari matriks identitas c adalah : {np.linalg.det(c)}")
print(f"determinan dari matriks identitas d adalah : {np.linalg.det(d)}")
print()
```

```python
a = np.array([[3,3], [5,2]])
B = np.linalg.inv(a)
detA = np.linalg.det(a)
detB = np.linalg.det(B)
print(f'determinan A : {detA.round()}')
print(f'determinan A invers : {detB.round()}')
C = detA * detB
print(f"det.A x det.B :",round(C))
print(a@B.round())
```
